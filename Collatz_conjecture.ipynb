```python
def whole():
    li = []
    temp = 0
    n = eval(input("Enter you number: "))
    li.append(n)
    while(True):
        if n%2 != 0:
            n = 3*n + 1
            li.append(n)
        else:
            n = n//2
            li.append(n)

        # statement to break the loop
        if n == 4:
            temp = 4
        if n == 2 and temp == 4:
            temp = 2
        if n == 1 and temp == 2:
            break
    print(*li)

while(True):
    whole()
```
